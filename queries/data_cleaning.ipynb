```python
import pandas as pd
import sqlite3

# CLEANING DATA

df = pd.read_csv("sql_data_clean.csv")

df.columns = [
    "year",
    "sex",
    "education_level",
    "employed_thousands"
]

conn = sqlite3.connect("sql_project.db")
df.to_sql("employed_by_education", conn, if_exists="replace", index=False)
conn.close()

print("Import successful!")
exit()
```
